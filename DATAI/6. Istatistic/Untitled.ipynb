```python
import numpy as np
from scipy import stats

yas = [100,13,44,23,56,13,68]
mean_yas = np.mean(yas)
median_yas = np.median(yas)
mode_yas = stats.mode(yas)

print(mean_yas)
print(median_yas)
print(mode_yas)
```

```python
yas = [100,13,44,23,56,13,68]
print("Range: ", (np.max(yas)- np.min(yas)))

print("Variance: ", np.var(yas))
var = sum((yas - np.mean(yas))**2) / len(yas)
print("Variance w formula: ", var)

print("Std: ", np.std(yas))
std = np.sqrt(var)
print("Std w formula: ", std)
```

```python
lamda = 3
s1 = np.random.poisson(lamda, 100000)
weights1 = np.ones_like(s1)/float(len(s1))
plt.hist(s1, weights=weights1, bins = 100)
plt.xlabel("number of accurances")
plt.ylabel("probability")
plt.show()
```

```python
mu, sigma = 110, 20
s = np.random.normal(mu, sigma, 100000)
print("mean: ", np.mean(s))
print("standart deviation: ", np.std(s))

plt.figure(figsize = (10, 7))
plt.hist(s, 100)
plt.xlabel("frequency")
plt.ylabel("IQ")
plt.title("Histogram of IQ")
plt.show()
```

```python
x = np.random.random_integers(10, size=100000)
plt.hist(x)
plt.show()
```

```python
import random
mean_sample = []
for i in range(10000):
    sample = random.randrange(5, 10)
    mean_sample.append(np.mean(random.sample(list(x), sample)))
plt.hist(mean_sample, bins = 10, color="red")
plt.show()
```

```python
plt.hist(x, alpha=0.5, density=True)
plt.hist(mean_sample, bins=50, alpha=0.5, color="red", density=True)
plt.title("Central Limit Theorem")
plt.show()
```
